WordNet Lemmatizer - it uses morphy() to wordnet corpus reader class to find  a lemma

````python
import nltk
nltk.download('wordnet')
````

````python
#chatbots,text summarizing
from nltk.stem import WordNetLemmatizer
````

````python
lemma=WordNetLemmatizer()
lemma.lemmatize("playing",pos='r')
````

````python
lemma.lemmatize("his",pos='n')
````
