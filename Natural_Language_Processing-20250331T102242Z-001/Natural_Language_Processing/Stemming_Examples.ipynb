  Stemming:Process of reducing words into its base form or root form

````python
#classsification using nlp
#sentimental analysis
#based on comments we gonna find whether it is positive or negative review
words=["Eaten","Programming","Going","Typing","Writing","finally","energetic"]
````

Porter Stemmer-Reducing the words into base form by removing its suffixes

```
# This is formatted as code
```



````python
from nltk.stem import PorterStemmer
````

````python
stemmer = PorterStemmer()
for word in words:
  print(word+"---->"+stemmer.stem(word))
````

RegExp Stemmer Class- It has single regular expression and removes prefixes or suffixes that matches that expression

````python
from nltk.stem import RegexpStemmer
reg_stem=RegexpStemmer('ing|ed$|s$|es$|able$|',min=2)
````

````python
words=["eating","knowledgable","besides","pumped","knows"]
for word in words:
  print(word+"---->"+reg_stem.stem(word))
````

````python
reg_stem.stem("ingsaravanaing")
````

Snow Ball Stemmer--->

````python
from nltk.stem import SnowballStemmer
snow=SnowballStemmer('english')
````

````python
for word in words:
  print(word+"-->"+snow.stem(word))
````
