Removing stopwords after applying PorterStemmer

```python

stemmmer=PorterStemmer()
```

```python
for i in range(len(sentences)):
  words=nltk.word_tokenize(sentences[i])
  no_stopwords=[stemmmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
  sentences[i]=' '.join(no_stopwords)
  print(sentences[i])
```
